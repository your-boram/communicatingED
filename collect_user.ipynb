```python
import pandas as pd

from twitter import Twitter

twitter = Twitter("edtwt", "2022-01-01", "2022-04-14")

df = pd.read_csv("twitter.csv")
urls = []
for row in df.itertuples():
    url = row.url.split("/status/")[0]
    urls.append(url)


user_stats, failed = twitter.get_user_page(urls)
user_stats.head()
```

```python
user_stats2, failed2 = twitter.get_user_page(failed)
```

```python
user_stats = pd.concat([user_stats, user_stats2])
```

```python
df_ = user_stats.transpose()
df_
```

```python
df = pd.concat([df_.iloc[:,0:3].dropna(), df_.iloc[:,3:].dropna()], axis=0)
df.head()
```

```python
df["num_tweet"]=df["num_tweet"].str.replace(" Tweets", "")
df["num_tweet"]=df["num_tweet"].str.replace("Tweet", "")
```

```python
df["num_tweet"]=df["num_tweet"].str.replace("K","00")
df["num_tweet"]=df["num_tweet"].str.replace(",","")
df["num_tweet"]=df["num_tweet"].str.replace(".","")

df["num_tweet"].astype(int)
```

```python
df["num_following"]=df["num_following"].str.replace(",","").astype(int)
df["num_follower"]=df["num_follower"].str.replace(",","").astype(int)
```

```python
result_df = df.sort_values(["num_follower","num_tweet"], ascending=False)
result_df.head()
```

```python
result_df.to_csv("result.csv")
```
